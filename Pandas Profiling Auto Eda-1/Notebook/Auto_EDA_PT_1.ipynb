#Importing Pandas and Pandas Profiling with ProfileReport

```python
import pandas as pd
from pandas_profiling import ProfileReport
```

##Reading Csv File

```python
df=pd.read_csv('CardioGoodFitness.csv')
```

##Passing the dataframe to Profile Report and Generating EDA Report in HTML File

```python
profile = ProfileReport(df)
profile.to_file(output_file='report.html')
```
